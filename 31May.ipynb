```python
import cv2
import numpy as np
img1=cv2.imread('C:/Users/User/Image Processing 2023/Program Files/2bit2.png')
img2=cv2.imread('C:/Users/User/Image Processing 2023/Program Files/1bit1.png')
i_and=cv2.bitwise_and(img1,img2)
i_or=cv2.bitwise_or(img1,img2)
i_xor=cv2.bitwise_xor(img1,img2)
i_not=cv2.bitwise_not(img2)
cv2.imshow('bitwise or',i_or)
cv2.imshow('bitwise and',i_and)
cv2.imshow('bitwise xor',i_xor)
cv2.imshow('bitwise not',i_not)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
